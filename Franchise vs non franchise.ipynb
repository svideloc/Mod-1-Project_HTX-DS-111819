```python
#import stuff
import pandas as pd
import seaborn as sns
```

```python
#import csv files
top_seven_df = pd.read_csv('top 7 for each year - Sheet1.csv')
top_2016_movies_df = pd.read_csv('2016 gross  - Sheet1.csv')
top_2017_movies_df = pd.read_csv('2017 - Sheet1.csv')
top_2018_movies_df = pd.read_csv('2018 - Sheet1.csv')
top_2016_movies_df.drop(top_2016_movies_df.head(1).index, inplace = True)
```

```python
#calculate totals function
def calc_sum(df,col_name):
    df[col_name].astype(str)
    df[col_name] = df[col_name].str.replace(',', '')
    df[col_name] = df[col_name].str.replace('$', '')
    df[col_name] = df[col_name].astype(float)
    total = sum(df[col_name])
    return total
```

```python
#calc franchise total
franchise_2016_total = calc_sum(top_seven_df,'World_Box_Office_2016_f')
franchise_2017_total = calc_sum(top_seven_df,'World_Box_Office_2017_f')
franchise_2018_total = calc_sum(top_seven_df,'World_Box_Office_2018_f')

#calc non-franchise total
nfranchise_2016_total = calc_sum(top_seven_df,'World_Box_Office_2016_nf')
nfranchise_2017_total = calc_sum(top_seven_df,'World_Box_Office_2016_n7')
nfranchise_2018_total = calc_sum(top_seven_df,'World_Box_Office_2018_nf')
```
