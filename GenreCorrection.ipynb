For this notebook we are adding new columns to with a 1 if that genre applies to that column, and a 0 if it doesnt. we felt this was the best way to adjust the data here since the genre showed up as symbols

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('final_data_set_sam')
```

```python
df.head()
```

```python
if '28' in df['genre'][0]:
    print('yes')
```

```python
'24' in df['genre'][0] 
```

```python
df['isAction'] = df['genre'].apply(lambda x: '28' in x).astype(np.uint8)
df['isAdventure'] = df['genre'].apply(lambda x: '12' in x).astype(np.uint8)
df['isAnimation'] = df['genre'].apply(lambda x: '16' in x).astype(np.uint8)
df['isComedy'] = df['genre'].apply(lambda x: '35' in x).astype(np.uint8)
df['isCrime'] = df['genre'].apply(lambda x: '80' in x).astype(np.uint8)
df['isDocumentary'] = df['genre'].apply(lambda x: '99' in x).astype(np.uint8)
df['isDrama'] = df['genre'].apply(lambda x: '18' in x).astype(np.uint8)
df['isFamily'] = df['genre'].apply(lambda x: '10751' in x).astype(np.uint8)
df['isFantasy'] = df['genre'].apply(lambda x: '14' in x).astype(np.uint8)
df['isHistory'] = df['genre'].apply(lambda x: '36' in x).astype(np.uint8)
df['isHorror'] = df['genre'].apply(lambda x: '27' in x).astype(np.uint8)
df['isMusic'] = df['genre'].apply(lambda x: '10402' in x).astype(np.uint8)
df['isMystery'] = df['genre'].apply(lambda x: '9648' in x).astype(np.uint8)
df['isRomance'] = df['genre'].apply(lambda x: '10749' in x).astype(np.uint8)
df['isScify'] = df['genre'].apply(lambda x: '878' in x).astype(np.uint8)
df['isTVmovie'] = df['genre'].apply(lambda x: '10770' in x).astype(np.uint8)
df['isThriller'] = df['genre'].apply(lambda x: '53' in x).astype(np.uint8)
df['isWar'] = df['genre'].apply(lambda x: '10752' in x).astype(np.uint8)
df['isWestern'] = df['genre'].apply(lambda x: '37' in x).astype(np.uint8)
```
