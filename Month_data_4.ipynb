This notebook was used to compile the final set of data that was used, calculating a month out of the year and roi, and a few other tweaks to shore up the final data set.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('new_data_sam_dec5')
```

```python
df.head()
```

```python
df['release_date'] = df['release_date'].astype('datetime64[ns]') 
```

```python
type(df.release_date[0])
```

```python
df.loc[df['release_date'].isna(), 'release_date'] = '2018-11-09'
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df['month_year'] = df['release_date'].dt.month.astype(np.int)
months = {1: 'January', 
          2: 'February', 
          3: 'March',
          4: 'April',
          5: 'May',
          6: 'June',
          7: 'July',
          8: 'August',
          9: 'September',
          10: 'October',
          11: 'November',
          12: 'December'}
df['month_year'] = df['month_year'].apply(lambda x: months[x])
```

```python
df['month#'] = df['release_date'].dt.month.astype(np.int)
```

```python
df['roi'] = (df['worldwide_gross - budget'] / df.budget)*100
```
